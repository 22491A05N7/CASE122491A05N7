```python
import pandas as pd
import matplotlib.pyplot as cn
import seaborn as cse

ntr=pd.read_csv('heart.csv')
print(ntr)

#scatterchart
cse.scatterplot(x='age',y='bp',data=ntr,hue='gender')
cn.title('heart attack')
cn.xlabel('age')
cn.ylabel('bp')
cn.show()

#linechart
cse.lineplot(x='age',y='cholestrol',data=ntr,hue='gender')
cn.show()

#boxplot
cse.boxplot(x='target',y='value',data=ntr)
cn.show()

#countplot
color=['red','yellow']
cse.countplot(x='target',data=ntr,palette=color)
cn.show()
```
